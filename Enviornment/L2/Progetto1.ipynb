# Data Preparation

```python
# Carica i tuoi dati
# Supponendo che tu abbia due array NumPy:
# - noisy_data: array di forma (200000, 600) con i segnali rumorosi
# - clean_data: array di forma (200000, 600) con i segnali puliti corrispondenti

# Carica i dati (sostituisci con il tuo codice per caricare i dati)
# Esempio:
# noisy_data = np.load('noisy_data.npy').astype(np.float32)
# clean_data = np.load('clean_data.npy').astype(np.float32)

# Per questo esempio, generiamo dati casuali
data_size = 200000
input_dim = 600

np.random.seed(0)
clean_data = df_pure[:,1,:].astype(np.float32)
noisy_data = df_noise[:,1,:].astype(np.float32)

# Dividi i dati in training, validation e test set
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

train_size = int(train_ratio * data_size)
val_size = int(val_ratio * data_size)
test_size = data_size - train_size - val_size

noisy_train = noisy_data[:train_size]
clean_train = clean_data[:train_size]

noisy_val = noisy_data[train_size:train_size+val_size]
clean_val = clean_data[train_size:train_size+val_size]

noisy_test = noisy_data[train_size+val_size:]
clean_test = clean_data[train_size+val_size:]

# Crea DataLoader
batch_size = 256

train_dataset = TensorDataset(torch.from_numpy(noisy_train), torch.from_numpy(clean_train))
val_dataset = TensorDataset(torch.from_numpy(noisy_val), torch.from_numpy(clean_val))
test_dataset = TensorDataset(torch.from_numpy(noisy_test), torch.from_numpy(clean_test))

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
test_loader = DataLoader(test_dataset, batch_size=batch_size)
```

# Model Initialization

```python
# Definisci l'autoencoder profondo
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(600, 512),
            nn.ReLU(True),
            nn.Linear(512, 256),
            nn.ReLU(True),
            nn.Linear(256, 128),
            nn.ReLU(True),
            nn.Linear(128, 64),
            nn.ReLU(True),
            nn.Linear(64, 32)
        )
        self.decoder = nn.Sequential(
            nn.Linear(32, 64),
            nn.ReLU(True),
            nn.Linear(64, 128),
            nn.ReLU(True),
            nn.Linear(128, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, 600)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

# Training Setup

```python


# Inizializza il modello, l'ottimizzatore, lo scheduler e la funzione di perdita
model = Autoencoder()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5)

# Parametri per l'early stopping
early_stopping_patience = 10
best_val_loss = float('inf')
epochs_no_improve = 0
num_epochs = 100


# Inizializza le liste per salvare le loss
train_losses = []
val_losses = []
# Ciclo di training
for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    for noisy_batch, clean_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(noisy_batch)
        loss = criterion(outputs, clean_batch)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    train_loss /= len(train_loader)
    train_losses.append(train_loss)  # Salva la loss di training

    # Validazione
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for noisy_batch, clean_batch in val_loader:
            outputs = model(noisy_batch)
            loss = criterion(outputs, clean_batch)
            val_loss += loss.item()
    val_loss /= len(val_loader)
    val_losses.append(val_loss)  # Salva la loss di validazione

    # Aggiorna lo scheduler
    scheduler.step(val_loss)

    print(f'Epoca {epoch+1}/{num_epochs}, Loss Training: {train_loss:.9f}, Loss Validazione: {val_loss:.9f}')

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_no_improve = 0
        best_model = model.state_dict()
    else:
        epochs_no_improve += 1
        print('No improvement in validation loss: {epochs_no_improve} times')
        if epochs_no_improve == early_stopping_patience:
            print('Early stopping!')
            break
```

```python
# Plot delle loss di training e validazione
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Loss Training')
plt.plot(val_losses, label='Loss Validazione')
plt.xlabel('Epoche')
plt.ylabel('Loss MSE')
plt.title('Andamento della Loss durante il Training')
plt.legend()
plt.grid(True)
plt.show()
```

# Model Evaluation

```python
# Carica il miglior modello
model.load_state_dict(best_model)

# Test del modello
model.eval()
test_loss = 0
with torch.no_grad():
    for noisy_batch, clean_batch in test_loader:
        outputs = model(noisy_batch)
        loss = criterion(outputs, clean_batch)
        test_loss += loss.item()
test_loss /= len(test_loader)
print(f'Loss Test: {test_loss:.6f}')

# Se vuoi salvare il modello
torch.save(model.state_dict(), 'denoising_model.pth')
```
